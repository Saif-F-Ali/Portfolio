## EXTRACTING IMPORTANT TAGS

- `<div class="row countdown-item" id="row-index-140" style="padding-left: 10px;">`
- `<div class="info director">`
- `<a href="https://www.rottentomatoes.com/m/1018009-running_scared">Running Scared</a>`
- `<span class="subtle start-year">(1986)</span>`
- `<span class="tMeterScore" style="margin-right: 10px;">63%</span>`
- `<div class="countdown-index">#140</div>`
- `<div class="info critics-consensus">`
    

## 1. movie name list

```python
import re
```

```python
# finding tags with movie names
names = soup.find_all("a", href=re.compile("https://www.rottentomatoes.com/m/"))
```

```python
len(names)
```

```python
# selecting the target tag containning the text name
names = names[1::3]
names[:11]
```

```python
# extracting the text
names_list = [i.text for i in names]
names_list[:11]
```

```python
# checking the integrity
len(names_list)
```

## 2. movie year list

```python
years = soup.find_all("span", class_="subtle start-year")
years[:4]
```

```python
len(years)
```

```python
years_list = [i.text.strip("()") for i in years]
years_list[:11]
```

## 3. movie score list

```python
scores = soup.find_all("span", class_="tMeterScore")
scores[:4]
```

```python
len(scores)
```

```python
score_list = [i.text for i in scores]
score_list[:11]
```

## 4. directors list

```python
directors = soup.find_all("div", class_="info director")
directors[:4]
```

```python
len(directors)
```

```python
directors_list = [i.text for i in directors]
directors_list[:5]
```

```python
# removing unnecessary text
directors_list = [i.removeprefix("\nDirected By: ") for i in directors_list]
directors_list[:11]
```

## 5. critics consensus list

```python
consensus = soup.find_all("div", class_="info critics-consensus")
consensus[:5]
```

```python
len(consensus)
```

```python
# extracting text
con_list = [i.text for i in consensus]
con_list[:5]
```

```python
# removing unnecessary text
con_list = [i.removeprefix("\nCritics Consensus: ") for i in con_list]
con_list[:5]
```

```python
# removing terminal spaces
con_list = [i.strip() for i in con_list]
con_list[:5]
```

## Creating the DataFrame

```python
rottentom_ActionMovies = pd.DataFrame()
```

```python
rottentom_ActionMovies["movie_name"] = names_list
rottentom_ActionMovies["movie_year"] = years_list
rottentom_ActionMovies["movie_director"] = directors_list
rottentom_ActionMovies["movie_score"] = score_list
rottentom_ActionMovies["critics_consensus"] = con_list
```
